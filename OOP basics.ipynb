```python
# Object Oriented programming~project 1

class Student:
    def __init__(self, name, age, grade):  # constructor
        self.name = name
        self.age = age
        self.grade = grade   # 0 - 100

    def get_grade(self):
        return self.grade


class Course:
    def __init__(self, name, max_students):
        self.name = name
        self.max_students = max_students
        self.students = []

    def add_student(self, student):
        if len(self.students) < self.max_students:
            self.students.append(student)
            return True
        return False

    def get_average_grade(self):
        total = sum(student.get_grade() for student in self.students)
        return total / len(self.students) if self.students else 0


# Create students
s1 = Student("Trick", 19, 98)
s2 = Student("Munyi", 18, 87)
s3 = Student("Maina", 21, 67)
s4 = Student("Lenny", 24, 70)
s5 = Student("Lorna", 25, 58)
s6 = Student("Bill", 23, 68)
s7 = Student("Jill", 21, 56)
s8 = Student("Kamau", 21, 54)
s9 = Student("Amad", 23, 90)
s10 = Student("Okoko", 23, 56)
s11 = Student("Rono", 23, 89)
s12 = Student("Yoro", 21, 78)
s13 = Student("Bryan", 34, 56)
s14 = Student("Kamau", 23, 90)

# Create course
course = Course("Mathematics", 2)
course.add_student(s1)
course.add_student(s2)

# Print first student's name
print(course.students[0].name)

# Print average grade
print("Average grade:", course.get_average_grade())
```
